### 1.1 Plotting number of accidents in the different age groups old,middle and young 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dataa = pd.read_csv('is-2018.csv')  
datab = pd.read_csv('is-2019.csv') 
datac = pd.read_csv('is-2020.csv') 
datad = pd.read_csv('is-2021.csv') 


df_younga = dataa[(dataa['age'] >= 15)&(dataa['age']<=24)] 
data_younga = pd.DataFrame(df_younga).to_numpy()

df_youngb = datab[(datab['age'] >= 15)&(datab['age']<=24)] 
data_youngb = pd.DataFrame(df_youngb).to_numpy()

df_youngc = datac[(datac['age'] >= 15)&(datac['age']<=24)] 
data_youngc = pd.DataFrame(df_youngc).to_numpy()

df_youngd = datad[(datad['age'] >= 15)&(datad['age']<=24)] 
data_youngd = pd.DataFrame(df_youngd).to_numpy()

#########################################################

df_middlea = dataa[(dataa['age'] >= 50)&(dataa['age']<=60)] 
data_middlea = pd.DataFrame(df_middlea).to_numpy()

df_middleb = datab[(datab['age'] >= 50)&(datab['age']<=60)] 
data_middleb = pd.DataFrame(df_middleb).to_numpy()

df_middlec = datac[(datac['age'] >= 50)&(datac['age']<=60)] 
data_middlec = pd.DataFrame(df_middlec).to_numpy()

df_middled = datab[(datab['age'] >= 50)&(datab['age']<=60)] 
data_middled = pd.DataFrame(df_middleb).to_numpy()
##################################################################
df_olda = dataa[dataa['age']>60]
data_olda = pd.DataFrame(df_olda).to_numpy()

df_oldb = datab[datab['age']>60]
data_oldb = pd.DataFrame(df_oldb).to_numpy()

df_oldc = datac[datac['age']>60]
data_oldc = pd.DataFrame(df_oldc).to_numpy()

df_oldd = datad[datad['age']>60]
data_oldd = pd.DataFrame(df_oldd).to_numpy()

plt.figure(figsize=(12,14))
x = ['2018', '2019', '2020', '2021']
y1 = np.array([len(df_younga), len(df_youngb), len(df_youngc), len(df_youngd)])
y2 = np.array([len(df_middlea) , len(df_middleb), len(df_middlec), len(df_middled)])
y3 = np.array([len(df_olda), len(df_oldb), len(df_oldc), len(df_oldd)])

# Plot bars in stack manner
plt.bar(x, y1, color='orange', width=0.5)
plt.bar(x, y2, bottom=y1, color='green', width=0.5)
plt.bar(x, y3, bottom=y1+y2, color='maroon',width=0.5)
plt.xlabel("Year")
plt.ylabel("Number of accidents")
plt.legend(("Young", "Middle", "Old"), loc=2)
plt.title("Number of accidents by age from 2018-2021")
plt.show()
```

### 1.3 Plotting number of accidents vs time in the respective years

```python
%reset
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import datetime


#import csv containing all date and places collated
#,format='%d%m%Y'

adateplace = pd.read_csv('adateplace.csv')  

adateplace['adate18'] = pd.to_datetime(adateplace['adate18'])
adateplace['adate19'] = pd.to_datetime(adateplace['adate19'])
adateplace['adate20'] = pd.to_datetime(adateplace['adate20'])
adateplace['adate21'] = pd.to_datetime(adateplace['adate21'])

# Extract adate count data and convert into dataframe
adatecount_18 = adateplace['adatecount18'].astype(int)
adatecount_19 = adateplace['adatecount19'].astype(int)
adatecount_20 = adateplace['adatecount20'].astype(int)
adatecount_21 = adateplace['adatecount21'].astype(int)

# Convert dataframe to a numpy array
dfcount18=pd.DataFrame(adatecount_18).to_numpy()
dfcount19=pd.DataFrame(adatecount_19).to_numpy()
dfcount20=pd.DataFrame(adatecount_20).to_numpy()
dfcount21=pd.DataFrame(adatecount_21).to_numpy()

#Plot number of accidents for the respective years
plt.figure(figsize=(15,18))
plt.subplot(5,1,1)
plt.plot(dfcount18[:,12],dfcount18[1:,13],"r",label="2018") 
plt.plot(dfcount19[1:,8],dfcount19[1:,9],"g",label="2019")
plt.plot(dfcount20[:,4],dfcount20[:,5],"b",label="2020")
plt.plot(dfcount21[:,0],dfcount21[:,1],"y",label="2021")

plt.title("Number of accidents for 2018-2021")
plt.ylabel("Number of accidents"); plt.xlabel("Date")
plt.legend()
plt.show()
```
